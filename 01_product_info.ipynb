# Geoscience Australia's Sentinel-1 IW Normalised Radar Backscatter (collection 0)

Geoscience Australia's Digital Earth (DE) branch are currently offering a suite of experimental Sentinel-1 normalised radar backscatter (NRB) products that we are calling collection 0, with sample data available over Australia and Antarctica.

The current offering contains data from [Sentinel-1's Interferometric Wide swath (IW) product](https://sentiwiki.copernicus.eu/web/s1-products).
The data have been processed using a [fork](https://github.com/GeoscienceAustralia/RTC) of NASA's [OPERA](https://www.earthdata.nasa.gov/data/projects/opera) Synthetic Aperture Radar [processing pipeline](https://github.com/opera-adt/RTC).
The data aim to meet the standards set out in the [CEOS-ARD Product Family Specification for Synthetic Aperture Radar](https://ceos.org/ard/files/PFS/SAR/v1.2/CEOS-ARD_PFS_Synthetic_Aperture_Radar_v1.2.pdf).

The product is a collaborative effort from Digital Earth Australia and Digital Earth Antarctica.
If you have any questions, please reach out to the Digital Earth Antarctica team at digitalearthantarctica@ga.gov.au.

## Data structure

A [Sentinel-1 IW acquisition](https://sentiwiki.copernicus.eu/web/s1-mission#S1Mission-InterferometricWideSwathS1-Mission-Interferometric-Wide-Swath) is composed of one 250 km swath, with three sub-swaths (IW1, IW2, IW3), with each sub-swath containing a number of bursts.
This is shown in the Figure below, with each sub-swath labelled. 
The bursts are visible as rectangles within each sub-swath.

![A Sentinel-1 acquisition showing three sub-swaths made up of multiple bursts.](https://sentiwiki.copernicus.eu/__attachments/1672913/IW_SLC_Product.jpg?inst-v=c9623a31-c9cf-47c0-be67-e455bd223601)

Bursts overlap almost perfectly between different passes and are always located at the same place, whereas scenes are not guarenteed to overlap.
This makes them an ideal data unit for repeatable analysis over time. 
As such, GA's Sentinel-1 Normalised Radar Backscatter product is delivered as bursts, rather than acquisitions/scenes.



## Differences between Australia and Antarctica

Sentinel-1 does not capture all polarisation modes globally.
Instead, horizontal-transmitted polarisation modes (HH and HV) are generally done over polar and ocean regions, and vertical-transmitted polarisation modes (VV and VH) are done everywhere else.
Differences in polarisation are due to [Sentinel-1's observation scenarios](https://sentinels.copernicus.eu/documents/247904/3944045/Sentinel-1-Mode-Polarisation-Observation-Geometry-2019.jpeg).
The table below captures the primary differences in the GA collection 0 offering:

| Property | Australia | Antarctica |
| --- | ----------- | ----------- |
| Primary capture mode | IW Vertical Dual-Polarisation (VV+VH) | IW Horizontal Single-Polarisation (HH) |
| Primary product | `ga_s1_nrb_iw_vv_vh_0` | `ga_s1_nrb_iw_hh_0` |
| Recommended CRS (metres) | UTM or EPSG:3577 | EPSG:3031 |
| Group by method | solar day | absolute orbit |

In Australia, there are a small number of IW Horizontal and Vertical Single-Polarisation acquisitions over south-east Queensland, but neither are the primary capture mode, and the collection 0 product only contains some data for this region.

## Available measurements

The normalised radar backscatter (NRB) measurements vary between products based on observing mode, and are provided in gamma0 convention.
Products also come with ancillary layers, known as static layers, which depend only on the observation geometry. 
The table below provides a summary of available measurements and the products they're available for.

| Measurement | Layer Type | Data Type | Units | No-data | Aliases | Products | Description |
| ----------- | ---------- | --------- | ----- | ------- | ------- | -------- | ----------- |
| VV_gamma0 | NRB | float32 | Linear Power | NaN | vv_gamma0 | `ga_s1_nrb_iw_vv_0`, `ga_s1_nrb_iw_vv_vh_0` | Co-polarised vertical backscatter in gamma0 convention. |
| VH_gamma0 | NRB | float32 | Linear Power | NaN | vh_gamma0 | `ga_s1_nrb_iw_vv_vh_0` | Cross-polarised vertical backscatter in gamma0 convention. |
| HH_gamma0 | NRB | float32 | Linear Power | NaN | hh_gamma0 | `ga_s1_nrb_iw_hh_0` | Co-polarised horizontal backscatter in gamma0 convention.|
| HV_gamma0 | NRB | - | - | - | - | Not available | Cross-polarised horizontal backscatter in gamma0 convention. |
| mask | Pixel quality | uint8 | Classification | 255 | - | All | A per-pixel layer indicating data validity, and whether the pixel has been affected by layover, shadow, or both. Classification values: `0` = valid; `1` = valid affected by shadow; `2` = valid affected by layover; `3` = valid affected by layover and shadow; `255` = invalid. |
| gamma0_to_beta0_ratio | Static | float32 | Ratio | NaN | - | All | A per-pixel conversion layer to convert from gamma0 to beta0. |
| gamma0_to_sigma0_ratio | Static | float32 | Ratio | NaN | - | All | A per-pixel conversion layer to convert from gamma0 to sigma0. |
| number_of_looks | Static | float32 | Count | NaN | n_looks, nlooks | All | A per-pixel layer that indicates the number of radar samples used in multi-looking during processing. | 
| incidence_angle | Static | float32 | Degrees | NaN | ia, IA | All | A per-pixel layer that gives the angle between the line-of-sight vector and the ellipsoid normal vector. | 
| local_incidence_angle | Static | float32 | Degrees | NaN | lia, LIA | All | A per-pixel layer that gives the angle between the line-of-sight vector and the terrain normal vector. |

## Available metadata

The table below provides a summary of several key pieces of metadata that may be useful as part of loading and filtering data.

| Dataset Metadata | Datacube Name | STAC Name | Description | Example Value |

| Scene ID | scene_id | sarard:scene_id | The name of the  scene/acquistion that the dataset belongs to. |  S1A_IW_SLC__1SDV_20201129T192619_20201129T192647_035467_042557_D8B8 |
| Sub-swath ID | beam_id | sarard:beam_id | The name of the sub-swath that the dataset belongs to. | IW1 |
| Burst ID | burst_id | sarard:burst_id | The burst identifier for the dataset. | t045_095837_iw1 |
| Relative orbit | relative_orbit | sat:absolute_orbit | The orbit path number, also known as the track. Each repeat pass over the same area has the same relative orbit. | 45 |
| Absolute orbit | absolute_orbit | sat:absolute_orbit | The number of times the satellite has travelled this orbit since launch, making it unique to the current orbit. | 35467 |
| Orbit state | orbit_state | sat:orbit_state | Whether the satellite is travelling north to south (descending) or south to north (ascending). | descending |
